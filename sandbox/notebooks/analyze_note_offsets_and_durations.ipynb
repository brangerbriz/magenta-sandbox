```python
import sys
sys.path.append('../python/')

import os
import data_utils
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12, 8)
```

```python
midi_dir = '../../data/query_symlinks'
```

```python
files = [os.path.join(midi_dir, path) for path in os.listdir(midi_dir)][0:400]

train_generator = data_utils.get_data_generator(files, 
                                                form=data_utils.F_EVENT_WINDOW_TIMES, 
                                                window_size=20,
                                                batch_size=1000)
X, y = train_generator.next()
```

```python
print(y)
```

```python
a = [n[0] for n in y]
b = [n[1] for n in y]
plt.scatter(a, b, alpha=0.03)
plt.xlim(0, 1)
plt.ylim(0, 2)
plt.show()
```

```python
print('Offset mean: {:.3f}\nDuration mean: {:.3f}'.format(np.mean(a), np.mean(b)))
print('Offset stddev: {:.3f}\nDuration stddev: {:.3f}'.format(np.std(a), np.std(b)))
```
